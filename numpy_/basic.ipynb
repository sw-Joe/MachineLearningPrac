```python
import torch
import numpy as np
```

```python
# 텐서 초기화
data = [[1, 2], [3, 4]]
xdata = torch.tensor(data)
print(xdata)
print(xdata.ndim)
```

```python
# 다른 텐서로부터 생성
x_ones = torch.ones_like(xdata)    # 속성(모양, 자료형) 유지
print(f"Ones Tensor: \n {x_ones} \n")

x_rand = torch.rand_like(xdata, dtype=torch.float) # 속성 오버라이드
print(f"Random Tensor: \n {x_rand} \n")
```

```python
# 무작위 또는 상수 값을 사용
shape = (2, 3)
rand_tensor = torch.rand(shape)    # 랜덤한 값의 행렬
ones_tensor = torch.ones(shape)    # 1행렬(항등행렬)
zeros_tensor = torch.zeros(shape)    # 0행렬

print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor}")
```

```python
# 속성
tensor = torch.rand(3, 4)

print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

```python
# 텐서 연산
# GPU가 존재하면 텐서를 이동합니다
if torch.cuda.is_available():
  tensor = tensor.to('cuda')
  print(f"Device tensor is stored on: {tensor.device}")
```

```python
# numpy식 표준 인덱싱, 슬라이싱
tensor = torch.ones(4, 4) 
tensor[:,1] = 0
print(tensor)
```

```python
# 텐서 합치기
t1 = torch.cat([tensor, tensor, tensor], dim=1)
t1
```

```python
# 텐서 곱(요소별 곱)
print(tensor.mul(tensor))    # tensor * tensor
```

```python
tensor.T
```

```python
# 두 텐서 간 행렬 곱(matrix multiplication)
print(tensor.matmul(tensor.T))    # tensor # tensor.T 
```

```python
# 바꿔치기(in-place 연산, '_' 접미사를 갖음)
print(tensor, "\n")
tensor.add_(5)
print(tensor)

# 바꿔치기 연산은 메모리를 일부 절약하지만, 기록(history)이 즉시 삭제되어 도함수(derivative) 계산에 문제가 발생할 수 있습니다. 
#   따라서, 사용을 권장하지 않습니다.
```

#### tensor with NumPy

```python
t = torch.ones(5)
print(f"t: {t}")
n = t.numpy()
print(f"n: {n}")
```

```python
t.add_(1)
print(f"t: {t}")
print(f"n: {n}")
```

```python
n = np.ones(5)
t = torch.from_numpy(n)

np.add(n, 1, out=n)
print(f"t: {t}")
print(f"n: {n}")
```
